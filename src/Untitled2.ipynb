```python
import os
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# --- Налаштування шляхів ---
# Директорія, де збережено ваші матриці
LOGS_DIR = 'C:\\Data\\logs'

def show_saved_confusion_matrices(log_dir):
    """
    Знаходить усі файли 'confusion_matrix_*.png' у вказаній директорії
    та відображає їх.
    """
    print(f"--- Пошук матриць невідповідностей у {log_dir}/ ---")
    
    # Складаємо шаблон для пошуку
    search_pattern = os.path.join(log_dir, "confusion_matrix_*.png")
    
    # Знаходимо всі файли, що відповідають шаблону
    matrix_files = glob.glob(search_pattern)
    
    if not matrix_files:
        print(f"Не знайдено файлів матриць за шаблоном '{search_pattern}'.")
        print("Переконайтеся, що ви запустили скрипт тренування, і файли збереглися.")
        return

    print(f"Знайдено {len(matrix_files)} матриць. Відображення...")
    
    # Проходимо по кожному знайденому файлу
    for filepath in matrix_files:
        try:
            # Зчитуємо зображення
            img = mpimg.imread(filepath)
            
            # Отримуємо назву моделі з імені файлу для заголовка
            filename = os.path.basename(filepath)
            model_name = filename.replace('confusion_matrix_', '').replace('.png', '').replace('_', ' ').title()
            
            # Відображаємо зображення
            plt.figure(figsize=(10, 8))  # Встановлюємо розмір
            plt.imshow(img)             # Показуємо зображення
            plt.title(f"Матриця для моделі: {model_name}")
            plt.axis('off')             # Вимикаємо осі (вони вже є на зображенні)
            plt.show()                  # Показуємо графік
            
        except Exception as e:
            print(f"Не вдалося завантажити або відобразити {filepath}: {e}")

# --- Запуск функції ---
if __name__ == "__main__":
    show_saved_confusion_matrices(LOGS_DIR)
```
