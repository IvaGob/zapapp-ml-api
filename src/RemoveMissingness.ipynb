```python
import sqlite3
import pandas as pd
import numpy as np

# 1. Підключення до бази даних
conn = sqlite3.connect("C:/Databases/charge_database_alt.sqlite")
```

```python
# 2. Завантаження таблиці ChargingSessions
df = pd.read_sql_query("SELECT * FROM ChargingSessions;", conn)

# 3. Перевірка пропущених значень
print("Пропущені значення до заповнення:")
print(df[['energy_consumed_kwh', 'charging_rate_kw', 'distance_driven_km']].isnull().sum())
```

```python
# 4. Заповнення пропусків медіаною
for col in ['energy_consumed_kwh', 'charging_rate_kw', 'distance_driven_km']:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
    print(f"{col}: заповнено медіаною = {median_value:.2f}")
```

```python
# 5. Перевірка результату
print("\nПропущені значення після заповнення:")
print(df[['energy_consumed_kwh', 'charging_rate_kw', 'distance_driven_km']].isnull().sum())
```

```python
# 6. (за бажанням) Оновити таблицю у базі
df.to_sql("ChargingSessions", conn, if_exists="replace", index=False)

conn.close()
```
