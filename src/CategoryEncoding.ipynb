```python
import sqlite3
import pandas as pd

# 1. Підключення до бази даних
conn = sqlite3.connect("C:/Databases/charge_database_alt.sqlite")
# Завантаження таблиць
charging_df = pd.read_sql_query("SELECT * FROM ChargingSessions;", conn)
users_df = pd.read_sql_query("SELECT * FROM Users;", conn)
```

```python
print("charger_type:", charging_df['charger_type'].unique())
print("user_type:", users_df['user_type'].unique())
```

```python
# One-Hot Encoding для charger_type
charging_encoded = pd.get_dummies(charging_df, columns=['charger_type'])

# One-Hot Encoding для user_type
users_encoded = pd.get_dummies(users_df, columns=['user_type'])

users_encoded.head()
```

```python
charging_encoded.to_sql("ChargingSessions_Encoded", conn, if_exists="replace", index=False)
users_encoded.to_sql("Users_Encoded", conn, if_exists="replace", index=False)

conn.close()
```
